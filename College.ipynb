```python
#imports
from ISLP import load_data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
college = load_data('College')
college
```

```python
#distribution descriptions
college.describe().iloc[:,:4]
```

```python
#Scatterplot matrix for Top10perc, Apps, and Enroll

# Selecting the desired columns
columns_to_plot = ['Top10perc', 'Apps', 'Enroll']
subset_df = college[columns_to_plot]

# Creating a scatterplot matrix
scatter_matrix = pd.plotting.scatter_matrix(subset_df, alpha=0.8, figsize=(10, 10), diagonal='hist')

# Adding axis labels and adjusting layout
plt.suptitle('Scatterplot Matrix of Top10perc, Apps, and Enroll', y=0.95, fontsize=16)
plt.tight_layout(rect=(0, 0, 1, 0.95))

# Show the plot
plt.show()
```

```python
# Boxplot of Outstate versus Private
college.boxplot(column='Outstate', by='Private', figsize=(8, 6))

# Adding labels and title
plt.title('Out-of-State Tuition for Private vs. Non-Private Colleges')
plt.xlabel('Private')
plt.ylabel('Out-of-State Tuition (dollars)')

# Show the plot
plt.show()
```

```python
#elite status

#add elite column
college['Elite'] = pd.cut(college['Top10perc'],[0,50,100], labels=['No', 'Yes'])

#count elite counts
value_counts = college['Elite'].value_counts()
print(value_counts)
```

```python
# Boxplot of Outstate versus Elite
college.boxplot(column='Outstate', by='Elite', figsize=(8, 6))

# Adding labels and title
plt.title('Out-of-State Tuition for Elite vs. Non-Elite Colleges')
plt.xlabel('Elite')
plt.ylabel('Out-of-State Tuition (dollars)')

# Show the plot
plt.show()
```
